```python
import matplotlib.pyplot as plt

# List of LaTeX equations
latex_equations = [
    r'S_x(f) = S_{\lambda}(f) + \lambda_0',
    r'\lambda(t) \sim {{Gaussian Process}}(\lambda_0, S_{\lambda})',
    r'X(t) \sim {{Poisson Process}}(\lambda(t))',
    r'|H^{(PSP)}(f)|^2',
    r'h^{(PSP)}(t)',
    r'H^{(LP)} = \frac{1}{\tau_{LP} \cdot f}',
    r'h^{(LP)}(t) = \frac{-t}{e^{\tau_{LP}}}',
    r'S_{\lambda}(f)',
    r'S_x(f) \cdot |H^{(PSP)}|^2 \cdot |H^{(LP)}|^2',
    r'X(t) \circledast h^{(PSP)}(t) \circledast h^{(LP)}(t)',
    r'\circledast'

]

latex_filenames = [
    'spectral_density_with_baseline',
    'gaussian_process_intensity',
    'poisson_process_gaussian_intensity',
    'psp_PSD',
    'psp_time_domain',
    'low_pass_filter_S_f',
    'low_pass_filter_X_t',
    'S_f',
    'Final_S_x',
    'final_X_t',
    'convolution_symbol'
]


def save_and_display_latex(latex_code, file_name, dpi=500):
    import os
    # Base directory path
    base_path = '/Users/patrick_bloniasz/filtered-point-process/filtered_point_process/Notebooks/Figures_Bloniasz_et_al_2024/Figure_2_Latex'

    # Check if the directory exists, if not create it
    if not os.path.exists(base_path):
        os.makedirs(base_path)

    # Set up a plot without axes
    fig, ax = plt.subplots()
    ax.axis('off')  # Hide axes

    # Set LaTeX text
    ax.text(0.5, 0.5, f'${latex_code}$', fontsize=100, ha='center', va='center')
    fig.tight_layout()

    # Save as a high-resolution PNG file
    png_file_path = os.path.join(base_path, f'{file_name}.png')
    fig.savefig(png_file_path, format="png", dpi=dpi, bbox_inches='tight', pad_inches=0.1, transparent=True)

    # Save as a high-resolution SVG file
    svg_file_path = os.path.join(base_path, f'{file_name}.svg')
    fig.savefig(svg_file_path, format="svg", dpi=dpi, bbox_inches='tight', pad_inches=0.1, transparent=True)

    print(f'Files saved at: {png_file_path} and {svg_file_path}')

    # Show the plot
    plt.show()
    plt.close(fig)

# Loop through and save each LaTeX equation with its descriptive filename
for equation, filename in zip(latex_equations, latex_filenames):
    save_and_display_latex(equation, filename)
```
