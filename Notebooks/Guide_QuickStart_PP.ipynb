## PP Quick Start Guide (Examples)


```python
import matplotlib.pyplot as plt
import os
import numpy as np

os.chdir('/Users/patrick_bloniasz/filtered_point_process')
```

```python
from filtered_point_process.PointProcess import PointProcessModel

model = PointProcessModel(seed=45)
```

```python
model = PointProcessModel(seed=45)
model.simulate_pp()  # Simulate the CIF model
print("CIF time domain time axis:", model.cif_time_axis)
print("CIF time domain intensity:", model.cif_realization)
print("CIF frequency domain frequencies:", model.cif_frequencies)
print("CIF frequency domain PSD:", model.cif_PSD)



# Create a figure with two subplots sharing the x-axis
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 8))

# Plot the CIF intensity realization
ax1.plot(model.cif_time_axis, model.cif_realization.squeeze(), color='blue')
ax1.set_ylabel('CIF Intensity')
ax1.set_title('CIF Intensity and Raster Plot of Point Process Spikes')

# Plot the raster plot of spikes
ax2.eventplot(model.pp_events, orientation='horizontal', colors='black', alpha = 0.3)
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Spikes')

# Show the combined plot
plt.tight_layout()
plt.show()
```
